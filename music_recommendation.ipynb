```python
import warnings
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm
import pickle
sns.set()
```

```python
data = pd.read_csv("dataset.csv")
data.head(5)
```

```python
data.info()
```

**removing null values**

```python
data.isnull()
```

```python
data.isnull().sum()
```

```python
data = data.dropna(axis = 0)
data.isnull().sum()
```

```python
data = data.drop_duplicates('track_name', keep='last')
data
```

**seprating numerical and object or string data**

```python
data.info()
```

```python
df_num = data.drop(columns = {'track_name','album_name','artists','track_id','explicit','track_genre'})
df_num.info()
```

**Finding co relation** 

```python
df_num.corr()
```

**normalizing** **values**

```python
from sklearn.preprocessing import MinMaxScaler
for col in df_num.columns:
  MinMaxScaler(col)
df_num.head()
```

**applying** **Kmeans** **algo**

```python
import joblib
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 10)
cluster = kmeans.fit_predict(df_num)
df_num['Cluster_no'] = cluster
MinMaxScaler(df_num['Cluster_no'])
```

```python
df_num['name'] = data['track_name']
df_num
```

```python

c1 = df_num[df_num['Cluster_no'] == 4]
c1
```

**Prediction**

```python
def abc(song , n):
  temp = df_num.loc[df_num['name'] == song, 'Cluster_no']
  print(temp)
  res = df_num[df_num['Cluster_no'] == int(temp)]
  res = res.result_df = res.drop_duplicates()
  print(res['name'].head(n))
  pass
```

```python
abc("Hunger" , 10)
```
